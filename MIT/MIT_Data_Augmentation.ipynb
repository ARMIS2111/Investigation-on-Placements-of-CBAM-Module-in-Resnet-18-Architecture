```python
# Path to the directory containing original images
dataset_path = '/content/indoorCVPR_09/Images'

# Create an ImageDataGenerator for data augmentation
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)
```

```python
# Create a directory for augmented images on Google Drive
!mkdir /content/drive/MyDrive/Datasets/MITIndoorDataset
```

```python
!mkdir /content/drive/MyDrive/Datasets/MITIndoorDataset/indoor_CVPR_09_augmented
```

```python
output_path = '/content/drive/MyDrive/Datasets/MITIndoorDataset/indoor_CVPR_09_augmented'
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".jpg") or file.endswith(".png"):
            # Load the image
            img_path = os.path.join(root, file)
            img = img_to_array(load_img(img_path, target_size=(224, 224)))  # Load and resize the image to 400x400

            # Create a directory for the augmented images if it doesn't exist
            category = os.path.basename(root)
            output_category_path = os.path.join(output_path, category)
            if not os.path.exists(output_category_path):
                os.makedirs(output_category_path)

            # Generate augmented images
            i = 0
            for batch in datagen.flow(np.expand_dims(img, axis=0), batch_size=1):
                augmented_img = array_to_img(batch[0])
                augmented_img.save(os.path.join(output_category_path, f'aug_{i}_{file}'))  # Save augmented image
                i += 1
                if i >= 5:  # Generate 5 augmented images for each original image
                    break  # Break the loop to prevent infinite generation
```
